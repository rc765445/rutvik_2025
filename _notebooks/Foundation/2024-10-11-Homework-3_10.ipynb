```python
# List management program
def add_element(my_list, element):
    my_list.append(element)
    print(f"{element} added to the list.")

def remove_element(my_list, element):
    if element in my_list:
        my_list.remove(element)
        print(f"{element} removed from the list.")
    else:
        print(f"{element} not found in the list.")

def check_element(my_list, element):
    if element in my_list:
        print(f"{element} exists in the list.")
    else:
        print(f"{element} does not exist in the list.")

def display_list(my_list):
    if my_list:
        print("Current list:", my_list)
    else:
        print("The list is empty.")

def find_max(my_list):
    if not my_list:
        return None
    max_value = my_list[0]
    for value in my_list:
        if value > max_value:
            max_value = value
    return max_value

def find_min(my_list):
    if not my_list:
        return None
    min_value = my_list[0]
    for value in my_list:
        if value < min_value:
            min_value = value
    return min_value

# Main program loop
def main():
    my_list = []
    while True:
        print("\nChoose an option:")
        print("1. Add element")
        print("2. Remove element")
        print("3. Check if element exists")
        print("4. Display list")
        print("5. Find max value")
        print("6. Find min value")
        print("7. Exit")
        
        choice = input("Enter your choice: ")
        
        if choice == '1':
            element = input("Enter element to add: ")
            add_element(my_list, element)
        elif choice == '2':
            element = input("Enter element to remove: ")
            remove_element(my_list, element)
        elif choice == '3':
            element = input("Enter element to check: ")
            check_element(my_list, element)
        elif choice == '4':
            display_list(my_list)
        elif choice == '5':
            max_value = find_max(my_list)
            if max_value is not None:
                print("Maximum value in the list:", max_value)
            else:
                print("The list is empty.")
        elif choice == '6':
            min_value = find_min(my_list)
            if min_value is not None:
                print("Minimum value in the list:", min_value)
            else:
                print("The list is empty.")
        elif choice == '7':
            print("Exiting program.")
            break
        else:
            print("Invalid choice, please try again.")

if __name__ == "__main__":
    main()


```

```python
// Shopping cart array
let shoppingCart = [];

// Function to display the shopping cart with icons
function displayCart() {
    if (shoppingCart.length === 0) {
        console.log("Your cart is empty!");
    } else {
        console.log("Your shopping cart contains:");
        console.log(shoppingCart.join(' ')); // Shows cart as icons (items)
    }
}

// Function to add an item to the cart
function addItem(item) {
    shoppingCart.push(item);
    console.log(`${item} has been added to your cart.`);
    displayCart();
}

// Function to remove an item from the cart
function removeItem(item) {
    const index = shoppingCart.indexOf(item);
    if (index !== -1) {
        shoppingCart.splice(index, 1);
        console.log(`${item} has been removed from your cart.`);
    } else {
        console.log(`${item} is not in your cart.`);
    }
    displayCart();
}

// Function to clear the cart
function clearCart() {
    shoppingCart = [];
    console.log("Your cart has been cleared.");
    displayCart();
}

// Main shopping actions
function startShopping() {
    addItem("üçé"); // Add an apple
    addItem("üçû"); // Add a bread
    addItem("ü•õ"); // Add milk
    removeItem("üçû"); // Remove bread
    addItem("üç™"); // Add a cookie
    clearCart(); // Clear the cart
}

// Start the shopping simulation
startShopping();
```
