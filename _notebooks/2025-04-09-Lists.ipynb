### Homework Hack #3


```python
import csv

def filter_spotify_data(file_path):
    # List to store filtered songs with more than 10 million streams
    filtered_songs = []

    # Open the CSV file
    with open(file_path, newline='', encoding='utf-8') as csvfile:
        # Create a CSV reader
        csvreader = csv.DictReader(csvfile)

        # Iterate through each row in the CSV file
        for row in csvreader:
            try:
                # Convert streams to an integer for comparison
                streams = int(row['Streams'])
                
                # Filter songs with more than 10 million streams
                if streams > 10000000:
                    filtered_songs.append(row)
            except ValueError:
                continue  # Skip rows with invalid stream values

    # Display the filtered data clearly
    print(f"Songs with more than 10 million streams:\n")
    for song in filtered_songs:
        print(f"Song: {song['Song Name']} | Artist: {song['Artist']} | Streams: {song['Streams']}")

```
